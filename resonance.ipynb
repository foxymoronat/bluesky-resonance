# Resonance — deine Bluesky-Interaktions-Feeds

Dieses Notebook baut dir **zwei Bluesky-Listen**, die du als Feeds anpinnen kannst:

- **Echo** — Accounts, deren Beiträge du oft **likest**
- **Amplify** — Accounts, die du oft **repostest oder zitierst**

Die Auswahl ist *zeitgewichtet*: jüngere Interaktionen zählen mehr als alte.

## Was du brauchst
Ein **App-Passwort** (nicht dein echtes Passwort!): in Bluesky unter *Settings → Privacy and Security → App Passwords → Add App Password*. Es ist jederzeit widerrufbar.

## So geht's
Die Zellen der Reihe nach mit dem ▶-Knopf ausführen (oder oben *Laufzeit → Alle ausführen*). Es wird **nichts auf deinem Gerät installiert** — alles läuft auf Googles Servern. Dein App-Passwort wird nur kurz abgefragt und **nirgends gespeichert**.


## Schritt 1 — Programm laden

Lädt die Bibliothek `atproto` und das Skript `resonance.py`. (Falls eine Meldung zum *Neustart der Laufzeit* erscheint: neu starten und diese Zelle erneut ausführen.)

```python
!pip install -q atproto
!wget -q -O resonance.py https://raw.githubusercontent.com/foxymoronat/bluesky-resonance/main/resonance.py
import resonance
print('Bereit! resonance.py geladen.')
```

## Schritt 2 — Anmelden & Vorschau

Hier gibst du dein Handle und App-Passwort ein. Diese Zelle macht eine **Vorschau** (Trockenlauf) — sie zeigt, was passieren würde, **schreibt aber nichts** in dein Konto.

```python
from getpass import getpass

handle = input('Dein Bluesky-Handle (ohne @), z.B. name.bsky.social: ').strip()
app_password = getpass('App-Passwort (bleibt unsichtbar): ').strip()

# Vorschau - schreibt NICHTS:
resonance.run(handle, app_password, 'dry-run')
```

## Schritt 3 — Listen wirklich erstellen

Sah die Vorschau gut aus? Dann legt diese Zelle die zwei Listen wirklich an (oder aktualisiert sie). Du kannst sie später jederzeit erneut ausführen, um die Feeds aufzufrischen — es entstehen keine Duplikate.

```python
# Erst ausfuehren, wenn die Vorschau oben gut aussah!
resonance.run(handle, app_password, 'write')
```

## Fertig! 🎉

Öffne Bluesky → linkes Menü **Listen** → *Echo Resonance* bzw. *Amplify Resonance* → **Pin to Home**. Danach erscheinen sie als Reiter oben auf deiner Startseite.
